```python
def cpu_details(df):
    intel_ind = df[df['CPU Brand']=='Intel'].index
    amd_ind = df[df['CPU Brand']=='AMD'].index

    df['CPU Series'] = df['CPU Name'].apply(lambda x: x.split(' ')[1] if len(x.split(' ')) > 2 else 'NA') # Athlon, Core, NA, Pentium, Ryzen, ore
    df.loc[amd_ind, 'CPU Series'] = df.loc[amd_ind, 'CPU Name'].apply(lambda x:
                                                                      ' '.join(x.split(' ')[1:3]) if len(x.split(' '))==5 else
                                                                      (' '.join(x.split(' ')[1:3]) if len(x.split(' '))==4 and 'Athlon' in x else
                                                                       (x.split(' ')[1] if len(x.split(' '))==4 and 'Athlon' not in x else 'NA'))) # Athlon 64, Athlon II, Athlon X2, Athlon X4, Ryzen

    df.loc[intel_ind, 'CPU Model'] = df.loc[intel_ind, 'CPU Name'].apply(lambda x:
                                                                         x.split(' ')[2] if x.split(' ')[2] in ['Dual-Core', 'Gold', 'Silver'] else
                                                                         (x.split(' ')[2].split('-')[0] if len(x.split(' ')[2].split('-'))>1 else 'NA')) # Dual-Core, Gold, NA, Silver, i3, i5, i7, i9
    ind = df[df['CPU Series']=='Ryzen'].index
    df.loc[ind, 'CPU Model'] = df.loc[ind, 'CPU Name'].apply(lambda x: x.split(' ')[-2] if x.split(' ')[-2]=='Threadripper' else x.split(' ')[-2][0])
    
    ind = df[df['CPU Series'].isin(['Athlon 64', 'Athlon II', 'Athlon X2', 'Athlon X4'])].index
    df.loc[ind, 'CPU Generation'] = df.loc[ind, 'CPU Name'].apply(lambda x: x.split(' ')[-1][0])
    df['CPU Model'] = df[df['CPU Series']=='NA']['CPU Name'].apply(lambda x: x.split(' ')[1].split('-')[0] if len(x.split(' '))==2 else None) # A4, A6, FX

    df['CPU Name'] = df['CPU Name'].apply(lambda x: x.replace('-', ' '))

    ind = df[df['CPU Brand']=='AMD'].index
    df.loc[ind, 'CPU Generation'] = df.loc[ind, 'CPU Name'].apply(lambda x: x.split(' ')[-1][0])

    return df
```

```python
df = cpu_details(df)
```
